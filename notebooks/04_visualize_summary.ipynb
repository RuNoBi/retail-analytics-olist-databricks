```python
df_summary = spark.table("puttaradol.runtest.customer_summary_payments")
display(df_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

panda = df_summary.toPandas()

plt.figure(figsize=(10,6))
sns.histplot(panda['total_spent'], bins=50, kde=True)
plt.title('Distribution of Total Spending per Customer')
plt.xlabel('TotalAmount')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

panda = df_summary.toPandas()

plt.figure(figsize=(10, 6))
sns.histplot(panda['avg_order_value'], bins=50, kde=True, color='orange')
plt.title("Distribution of Average Order Value")
plt.xlabel("Average Order Value")
plt.ylabel("Number of Customers")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

panda = df_summary.toPandas()

plt.figure(figsize=(10, 6))
sns.countplot(x='order_count', data=panda)
plt.title("Number of Orders per Customer")
plt.xlabel("Order Count")
plt.ylabel("Number of Customers")
plt.grid(True)
plt.show()
```
